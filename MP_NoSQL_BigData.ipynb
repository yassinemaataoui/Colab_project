## Partie 2 – Création du jeu de données JSON

Créons un jeu de données contenant des informations sur des étudiants et leurs notes.


```python
import json

etudiants = [
  {
    "id": 1,
    "nom": "Amina",
    "filiere": "Informatique",
    "notes": [
      {"matiere": "Big Data", "note": 15},
      {"matiere": "Programmation", "note": 17},
      {"matiere": "Mathématiques", "note": 14}
    ]
  },
  {
    "id": 2,
    "nom": "Youssef",
    "filiere": "Réseaux",
    "notes": [
      {"matiere": "Systèmes", "note": 12},
      {"matiere": "Réseaux", "note": 14},
      {"matiere": "Sécurité", "note": 13}
    ]
  },
  {
    "id": 3,
    "nom": "Sara",
    "filiere": "Informatique",
    "notes": [
      {"matiere": "Big Data", "note": 18},
      {"matiere": "IA", "note": 17},
      {"matiere": "Programmation", "note": 16}
    ]
  }
]

with open("etudiants.json","w") as f:
    json.dump(etudiants,f,indent=2)

print("Fichier JSON créé avec succès.")
```

## Partie 3 – Insertion des données dans la base NoSQL (Mongita)

Mongita fonctionne comme MongoDB mais stocke les données localement dans Colab, sans configuration réseau.


```python
from mongita import MongitaClientDisk
import json

client = MongitaClientDisk()
db = client["enseignement_db"]
collection = db["etudiants"]

collection.delete_many({})  # Nettoyage
with open("etudiants.json") as f:
    data = json.load(f)
collection.insert_many(data)
print("Insertion réussie dans la base Mongita locale.")
```

## Partie 4 – Requêtes d'exploration des données

```python
# Afficher tous les étudiants
print("Liste complète des étudiants :")
for doc in collection.find():
    print(doc)

# Afficher seulement le nom et la filière (projection manuelle en Python)
print("\nNoms et filières des étudiants :")
for doc in collection.find():
    print({"nom": doc.get("nom"), "filiere": doc.get("filiere")})

# Rechercher les étudiants d'une filière donnée
print("\nÉtudiants en Informatique :")
for doc in collection.find({"filiere": "Informatique"}):
    print(doc)

# Afficher les étudiants ayant une note > 15 en Programmation
print("\nÉtudiants ayant une note > 15 en Programmation :")
for doc in collection.find():
    for note in doc.get("notes", []):
        if note["matiere"] == "Programmation" and note["note"] > 15:
            print(f"{doc['nom']} : {note['note']}")

# Compter le nombre total d’étudiants
print("\nNombre total d’étudiants :", collection.count_documents({}))
```

## Partie 5 – Exportation du dataset vers PySpark

```python
import pandas as pd

# Extraire tous les documents (sans projection)
docs = list(collection.find())

# Supprimer le champ _id qui pose problème (ajout automatique par Mongita)
for doc in docs:
    if "_id" in doc:
        del doc["_id"]

# Exporter vers JSON
pd.DataFrame(docs).to_json("etudiants_export.json", orient="records")
print("Export vers etudiants_export.json terminé.")
```

## Partie 6 – Analyse avec PySpark

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import explode, col, avg

spark = SparkSession.builder.appName("MiniProjet_Etudiants").getOrCreate()
df = spark.read.json("etudiants_export.json")
df.printSchema()
df.show(truncate=False)
```

```python
df_explode = df.withColumn("note", explode(col("notes")))
result = df_explode.groupBy("filiere").agg(avg(col("note.note")).alias("moyenne_filiere"))
result.show()
```

### **Recherche d’étudiants d’une filière spécifique**

### **Recherche d’étudiants ayant une note élevée dans une matière donnée**

```python
# Afficher les étudiants ayant une note > 15 en Programmation
print("\nÉtudiants ayant une note > 15 en Programmation :")
for doc in collection.find():
    for note in doc.get("notes", []):
        if note["matiere"] == "Programmation" and note["note"] > 15:
            print(f"{doc['nom']} : {note['note']}")
```

Les étudiants ayant une note supérieure à 15 en Programmation ont été sélectionnés. Cette étape démontre comment extraire des informations conditionnelles à partir des données imbriquées.

### **Modification d’une note**

```python
for doc in collection.find():
    for note in doc.get("notes", []):
        if note["matiere"] == "Big Data":
          note["note"] = note["note"] + 1
          print(f"{doc['nom']} : {note['note']}")
```

Une note en Big Data a été augmentée de 1 point pour un étudiant donné. Cette opération illustre la capacité de la base NoSQL à mettre à jour des informations précises à l’intérieur de documents complexes.

### **Suppression d’un étudiant**

```python
# Supprimer le champ _id qui pose problème (ajout automatique par Mongita)
for doc in docs:
    if "_id" in doc:
        del doc["_id"]
```

Un étudiant ajouté par erreur a été supprimé. Cette manipulation montre comment gérer la suppression de données dans une base NoSQL.

### **Comptage des étudiants**

```python
# Compter le nombre total d’étudiants
print("\nNombre total d’étudiants :",collection.count_documents({}))
```

Le nombre total d’étudiants dans la base a été calculé. Cette étape fournit une vue d’ensemble sur le volume de données et permet de vérifier l’intégrité de la base.

### **Comptage des étudiants par filière**

```python
# Récupérer tous les étudiants
docs = list(collection.find())

# Dictionnaire pour compter par filière
compte_filiere = {}

for doc in docs:
    filiere = doc.get("filiere")
    if filiere in compte_filiere:
        compte_filiere[filiere] += 1
    else:
        compte_filiere[filiere] = 1

# Affichage du résultat
print("Nombre d'étudiants par filière :")
for filiere, nb in compte_filiere.items():
    print(f"{filiere} : {nb}")
```

Un dictionnaire Python a été utilisé pour compter le nombre d’étudiants dans chaque filière. Cette analyse permet d’avoir un aperçu rapide de la répartition des étudiants par spécialité.

### **Identification de l’étudiant ayant la meilleure moyenne**

```python
docs = list(collection.find())

best = None
best_avg = 0

for d in docs:
    notes = [n["note"] for n in d["notes"]]
    avg = sum(notes) / len(notes)

    if avg > best_avg:
        best_avg = avg
        best = d["nom"]

print(best, best_avg)
```

En parcourant les données, l’étudiant ayant la meilleure moyenne globale a été identifié. Cette étape montre comment combiner les notes de différentes matières pour obtenir un indicateur global de performance.

# **6. Exportation vers PySpark**

```python
spark = SparkSession.builder.appName("Etudiant").getOrCreate()
```

# **7. Analyse avec PySpark**

```python
df = spark.read.csv("etudiants_export.json",header=True,inferSchema=True)
```

```python
df.printSchema()
```

### **Affichage du schéma du DataFrame**

```python
df = spark.read.json("etudiants_export.json", multiLine=True)
df.printSchema()
```

### **Transformation explode**

```python
from pyspark.sql.functions import explode, col, avg

df_explode = df.select(
    "filiere",
    explode(col("notes")).alias("note")
)
df_explode.show(truncate=False)
```

### **Calcul de la moyenne des notes par filière**

```python
moyenne_filiere = df_explode.groupBy("filiere").agg(avg(col("note.note")).alias("moyenne_filiere")).show()
```

Après export des données vers PySpark, les notes ont été “dépliées” pour calculer la moyenne par filière. Cela permet de comparer les performances générales des différentes filières.

### **Calcul de la moyenne des notes par matière**

```python
moyenne_matiere = df_explode.groupBy(col("note.matiere")).agg(avg(col("note.note")).alias("moyenne_matiere"))

moyenne_matiere.show()
```

Chaque matière a été analysée afin de déterminer les performances moyennes des étudiants. Cette étape fournit des informations utiles sur les matières où les étudiants réussissent le mieux ou le moins bien.

### **Identification de la filière ayant la meilleure moyenne**

```python
result = df_explode.groupBy("filiere").agg(avg(col("note.note")).alias("moyenne_filiere")).orderBy(col("moyenne_filiere").desc()).limit(1).show(truncate=False)
```

Les moyennes par filière ont été comparées pour identifier la filière la plus performante. Cette information est utile pour le suivi académique et l’évaluation des filières.

### **Identification de la matière avec les meilleures performances**

```python
moyenne_matiere = df_explode.groupBy(col("note.matiere")).agg(avg(col("note.note")).alias("moyenne_matiere")).orderBy((col("moyenne_matiere")).desc()).limit(1).show()
```

En comparant les moyennes par matière, la matière où les étudiants obtiennent les meilleurs résultats a été mise en évidence. Cela permet d’orienter les efforts pédagogiques.

### **Calcul de la moyenne générale**

```python
moyenne_generale = df_explode.select(avg(col("note.note")).alias("moyenne_generale"))

moyenne_generale.show()
```

La moyenne globale de toutes les notes a été calculée pour donner une vue d’ensemble sur la performance du groupe d’étudiants. Cela constitue un indicateur clé pour évaluer la réussite globale.

### **Ajout d’une colonne “mention”**

```python
from pyspark.sql.functions import explode, col, when, avg

df_explode = df.select("nom","filiere",explode(col("notes")).alias("note"))

df_avg = df_explode.groupBy("nom").agg(avg(col("note.note")).alias("moyenne"))

df_mention = df_avg.withColumn(
    "mention",
    when(col("moyenne") > 16, "Excellent")
    .when(col("moyenne") >= 14, "Bien")
    .when(col("moyenne") >= 12, "Assez Bien")
    .otherwise("Passable")
).show()
```

Une colonne supplémentaire a été créée pour attribuer une mention à chaque étudiant selon sa moyenne (par exemple “Excellent” ou “Bien”). Cela facilite l’interprétation des performances et fournit un classement clair.

# **8. Requêtes SQL avec SparkSQL**



```python
df.createOrReplaceTempView("etudiants_sql")
```

Un tableau temporaire a été créé afin d'interroger les données avec SQL.

### **Étudiants de la filière “Informatique”**

```python
spark.sql("SELECT *  FROM etudiants_sql  WHERE filiere = 'Informatique' ").show()
```

Requête de filtrage simple.

### **Moyenne des notes par matière**

```python
spark.sql("SELECT n.matiere AS matiere, AVG(n.note) AS moyenne_matiere  FROM etudiants_sql LATERAL VIEW explode(notes) t AS n GROUP BY n.matiere").show()
```

Agrégation basée sur la valeur de la matière.

### **Filière ayant la meilleure moyenne**

```python
spark.sql("SELECT filiere, AVG(n.note) AS moyenne_filiere FROM etudiants_sql  LATERAL VIEW explode(notes) t AS n  GROUP BY filiere  ORDER BY moyenne_filiere DESC LIMIT 1").show()
```

Tri après agrégation.

### **Nombre total d'étudiants par filière**

```python
result = spark.sql("SELECT filiere, COUNT(*) AS nombre_etudiants FROM etudiants_sql GROUP BY filiere").show()
```

Comptage regroupé par filière.

Cette partie montre l’intégration entre SQL et Spark dans l’analyse Big Data
